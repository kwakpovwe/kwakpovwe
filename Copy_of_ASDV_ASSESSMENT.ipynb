```r
#Install the necessary libraries:

install.packages("tidyverse")
install.packages("ggplot2")
install.packages("psych")

```

```r
library(tidyverse)

# Replace the path with the correct file path or URL to your CSV file

Country_Disparities <- read_csv("/content/Country Disparities.csv", col_names = TRUE)



```

```r
# Name
names(Country_Disparities)

# Head
head(Country_Disparities)

# Tail
tail(Country_Disparities)

# Str
str(Country_Disparities)

# Summary
summary(Country_Disparities)
```

```r
#Take a look at the Dataset

View(Country_Disparities)
```

```r
#Display a summary of each column in the Dataset:

summary(Country_Disparities)
```

```r
#Do a comprehensive descriptive statistical analysis on the data:

library(psych)

# Descriptive statistics
summary(Country_Disparities)

# Histogram
hist(Country_Disparities$`Access to electricity`)

# Boxplot
boxplot(Country_Disparities$`Access to electricity`)

# Scatterplot
plot(Country_Disparities$`Access to electricity`, Country_Disparities$`Individuals using the Internet`)


```

```r
#PLOTTING WITH TWO VARIABLES

# Load necessary libraries
library(tidyverse)
library(psych)


# Create a bar plot
ggplot(Country_Disparities, aes(x = `Country Name`, y = `Access to electricity`)) + geom_bar(stat = "identity")

# Create a histogram with a specified bin width
ggplot(Country_Disparities, aes(x = `Access to electricity`)) + geom_histogram(binwidth = 5)

# Create a scatter plot with a color gradient
ggplot(Country_Disparities, aes(x = `Access to electricity`, y = `Individuals using the Internet`)) + geom_point() + scale_color_gradient(low = "blue", high = "red")
```
